```python
#encoding
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_columns = ['department','region','education','gender','recruitment_channel']
one_hot_enc = OneHotEncoder(handle_unknown='ignore')
train_encoded_df = one_hot_enc.fit_transform(train_df[categorical_columns]).toarray()
test_encoded_df = one_hot_enc.fit_transform(test_df[categorical_columns]).toarray()
```

```python
#checking outliers
numerical_columns = ['previous_year_rating', 'length_of_service', 'awards_won?','no_of_trainings','avg_training_score']
plt.figure(figsize=(5,10))
sns.boxplot(data=test_data[numerical_columns])
plt.title('Boxplot of Numerical Columns')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.xticks(rotation=45,fontsize=10)
plt.show()
```

```python
outlier_checking = ['previous_year_rating', 'length_of_service', 'awards_won?','no_of_trainings','avg_training_score']
Q1 = test_data[outlier_checking].quantile(0.25)
Q3 = test_data[outlier_checking].quantile(0.75)
IQR = Q3 - Q1

outliers_lower = test_data[outlier_checking] < (Q1 - 1.5 * IQR)
outliers_upper = test_data[outlier_checking] > (Q3 + 1.5 * IQR)

outliers_mild = outliers_upper | outliers_lower
print(test_data[outliers_mild])
```

```python
test_data[outlier_checking] = test_data[outlier_checking].clip(lower=Q1 - 1.5 * IQR, upper=Q3 + 1.5 * IQR,axis=1)
```

```python
x_train = train_data.drop(columns=['employee_id','avg_training_score'])
y_train = train_data['avg_training_score']
```

```python
x_test = test_data.drop(columns=['employee_id','avg_training_score'])
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Remove rows with NaN values in 'avg_training_score'
train_data = train_data.dropna(subset=['avg_training_score'])

# Recreate x_train and y_train after removing NaN values
x_train = train_data.drop(columns=['employee_id','avg_training_score'])
y_train = train_data['avg_training_score']

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

```python
# Get the column names from x_train
x_train_columns = x_train.columns

# Reorder the columns in x_test to match x_train
x_test = x_test[x_train_columns]

# Now you can predict using the reordered x_test
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(train_data['avg_training_score'], y_pred)
print('Accuracy:', accuracy)
print('Classification Report:')
print(classification_report(train_data['avg_training_score'], y_pred))
print('Confusion Matrix:')
print(confusion_matrix(train_data['avg_training_score'], y_pred))

```
