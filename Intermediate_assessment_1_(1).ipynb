**18.** Load the ny-flights dataset to Python

````python
import rpy2.robjects as robjects
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter
# Load the nycflights13 library in R
robjects.r('if(!require(nycflights13)) install.packages("nycflights13")')

# Load the nycflights13 dataset in Python
robjects.r('library(nycflights13)')

with localconverter(robjects.default_converter+pandas2ri.converter):
  flights_df=robjects.r('flights')

flights_df
````

**19.** Which airline ID is present maximum times in the dataset

````python
max_airline_id = flights_df['carrier'].value_counts().idxmax()

print(max_airline_id)

````
